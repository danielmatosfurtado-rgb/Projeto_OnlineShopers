```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
raw_shoppers = Path("../raw/dirty/online-shoppers-dirty.csv")
```

```python
df_shoppers = pd.read_csv(raw_shoppers, sep=";")
```

```python
df_shoppers.head()
```

```python
df_shoppers.columns
```

```python
df_shoppers.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_shoppers.shape
```

```python
df_shoppers.dtypes
```

```python
df_shoppers.columns = [col.strip().lower() for col in df_shoppers.columns]
```

```python
df_shoppers.columns
```

```python
categorical_cols = ['administrative','administrative-duration','productrelated duration','bouncerates','month','pagevalues','visitortype','weekend', 'revenue']
```

```python
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df_shoppers[col].unique())
```

```python
def clean_category(series: pd.Series) -> pd.Series:
    return (
        series
        .astype(str)
        .str.strip()
        .str.lower()
    )
```

```python
for col in categorical_cols:
    df_shoppers[col] = clean_category(df_shoppers[col])
```

```python
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df_shoppers[col].unique())
```

```python
df_shoppers.replace(
    {
        "?": np.nan,
        "unknown": np.nan,
        "nan": np.nan
        
    },
    inplace=True
)
```

```python
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df_shoppers[col].unique())
```

```python
df_shoppers['month'].replace(     
    {
        "02": 'feb',
        "05": 'may',
        "11": 'nov'
        
    },
    inplace=True
)

df_shoppers['visitortype'].replace(     
    {
        'returning visitor': 'returning_visitor',
        "returning": 'returning_visitor',
        "new": 'new_visitor',
        "new visitor":'new_visitor'     
    },
    inplace=True
)

df_shoppers['weekend'].replace(     
    {
        '0': 'false',
        "1": 'true',
        "yes": 'true',
        "no":'false'     
    },
    inplace=True
)

df_shoppers['revenue'].replace(     
    {
        '0': 'false',
        "1": 'true',
        "yes": 'true',
        "no":'false'     
    },
    inplace=True
)
```

```python
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df_shoppers[col].unique())
```

```python
numerical_cols = ['administrative','administrative-duration','productrelated duration','bouncerates','pagevalues']
```

```python
for var in numerical_cols:
    print(f"\nColuna: {var}")
    print(df_shoppers[df_shoppers[var].astype(str).str.count(r"\.") > 1][var])
```

```python
df_shoppers[numerical_cols] = df_shoppers[numerical_cols].replace(r"^(\d+\.\d+)\..*$", r"\1", regex=True).replace(r"[^\d.]", "", regex=True)
```

```python
df_shoppers[numerical_cols]= df_shoppers[numerical_cols].apply(pd.to_numeric)
```

```python
df_shoppers.dtypes
```

```python
df_shoppers.isna().sum()
```

```python
cols_missing = ["administrative", "administrative-duration", "month", "visitortype", "weekend", "revenue"]
```

```python
df_shoppers[cols_missing].isna().sum(axis=1).value_counts()
```

```python
numerical_cols_miss_value = ["administrative","administrative-duration"]
```

```python
for col in numerical_cols_miss_value:
    df_shoppers[col].fillna(df_shoppers[col].median(), inplace=True)
```

```python
categorical_cols_miss_value = [ "month", "visitortype", "weekend", "revenue"]
```

```python
for col in categorical_cols_miss_value:
    mode_value = df_shoppers[col].mode()

    if len(mode_value) > 0:
        df_shoppers[col].fillna(mode_value[0], inplace=True)
    else:
        df_shoppers[col].fillna("unknown", inplace=True)
```

```python
df_shoppers.isna().sum().sum()
```

```python
df_shoppers.duplicated().sum()
```

```python
df_shoppers.drop_duplicates(inplace=True)
```
